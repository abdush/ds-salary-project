```python
def simple_title(title):
    title_lower = title.lower()
    if 'data scientist' in title_lower:
        return 'data scientist'
    if 'data engineer' in title_lower:
        return 'data engineer'
    if 'data analyst' in title_lower:
        return 'analyst'
    if 'machine learning' in title_lower:
        return 'mle'
    if 'manager' in title_lower:
        return 'manager'
    if 'director' in title_lower:
        return 'director'
    return 'na'

def seniority(title):
    title_lower = title.lower()
    if any(word in title_lower for word in ['sr', 'senior', 'lead', 'principal', 'manager']):
        return 'senior'
    if any(word in title_lower for word in ['jr', 'junior']):
        return 'senior'
    return 'na'
```

```python
df['simple_title'] = df['Job Title'].apply(simple_title)
df.simple_title.value_counts()
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```

```python
#Fix states
#New Jersey -> NJ; Virginia -> VA; Oregon -> OR
def clean_state(state):
    state_lower = state.lower()
    if state_lower == 'new jersey':
        return 'NJ'
    if state_lower == 'virginia':
        return 'VA'
    if state_lower == 'oregon':
        return 'OR'
#     if (state_lower == 'united states' or state_lower == 'remote'):
#         return 'na'
    return state
    
df['state'] = df.state.apply(clean_state)
df.state.value_counts()
```

```python
#job desc length
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
```

```python
#Competitor count
df['competitors_cnt'] = df['Competitors'].apply(lambda x: 0 if x == '-1' else len(x.split(',')))
#df[['Competitors','competitors_cnt']]
```

```python
df.company_txt = df.company_txt.apply(lambda x: x.replace('\r', '').replace('\n', ''))
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
words = " ".join(df['Job Description'])

def punctuation_stop(text):
    """remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wc.generate(text)

plt.figure(figsize=[10,10])
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
